```python
import re

"""
A specialized tokenizer for math functions
"""
def math_tokenizer(text):

    # Not all necessary, but still used.
    math_functions = ['sin', 'cos', 'tan', 'log', 'ln', 'exp', 'sqrt', 'max', 'min', 'abs', 'floor', 'ceil', 'round', 'sum', 'prod']

    # some patterns
    patterns = [
        (r'\*\*', 'OP'),
        (r'[\+\-\*/\^=<>]', 'OP'),
        (r'[\(\)\[\]\{\}]', 'BRACKET'),
        (r'[a-zA-Z_][a-zA-Z0-9_]*', 'VAR'),
        (r'\s+', 'SPACE'),
    ]

    tokens = []
    index = 0

    while index < len(text):
        if (text[index].isdigit() or (text[index] == '.' and index + 1 < len(text) and text[index + 1].isdigit())):

            # If number is there, match it.
            number_match = re.match(r'\d+\.\d+|\d+', text[index:])

            if number_match:

                num_str = number_match.group(0)

                for digit in num_str:
                    tokens.append(digit)

                index += len(num_str)
                continue

        # If it isn't a number, try to see if it's a knwon regex pattern
        match_found = False
        for pattern, token_type in patterns:
            regex = re.compile(pattern)
            match = regex.match(text[index:])

            # if it is a known operation, add it
            if match:
                token_text = match.group(0)
                tokens.append(token_text)
                index += len(token_text)
                match_found = True
                break

        # worst case, just add the token
        if not match_found:
            tokens.append(text[index])
            index += 1

    return tokens
```

```python
from torch.utils.data import Dataset
from torchtext.data import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
from torchtext.transforms import ToTensor
import numpy as np



class MathDataset(Dataset):

    def __init__(self, dataset, split= "train", vocab = None, max_samples = 100000):

        # Take our samples, and decode them
        self.dataset = dataset[split].take(max_samples)
        self.samples = [(example[0].decode('utf-8'), example[1].decode('utf-8')) for example in tfds.as_numpy(self.dataset)]

        # If vocab not passed in, create it.
        if vocab is None:
            self.vocab = self.create_vocab()
        else:
            self.vocab = vocab

    def tokenize(self, text):
        tokens = math_tokenizer(text)
        return tokens

    def yield_tokens(self):
        for q, a in self.samples:
            yield self.tokenize(q)
            yield self.tokenize(a)

    def create_vocab(self):

        vocab = build_vocab_from_iterator(self.yield_tokens(), specials = ["<unk>", "<pad>", "<start>", "<end>"], min_freq = 1)

        # Default unknown
        vocab.set_default_index(vocab["<unk>"])

        # Forced digits (was a problem)
        for char in "0123456789.":
            if char not in vocab:
                vocab.append_token(char)

        # Forced symbols
        for symbol in "+-*/^=()[]{}":
            if symbol not in vocab:
                vocab.append_token(symbol)

        # Force spaces
        if " " not in vocab:
            vocab.append_token(" ")

        return vocab

    def __len__(self):

        return len(self.samples)

    def __getitem__(self, i):

        # Get question and answer tokens, append start
        question, answer = self.samples[i]
        q_tokens = [self.vocab["<start>"]]
        a_tokens = [self.vocab["<start>"]]

        # Add their tokens
        q_tokenized = self.tokenize(question)
        q_indices = self.vocab.lookup_indices(q_tokenized)
        q_tokens.extend(q_indices)

        a_tokenized = self.tokenize(answer)
        a_indices = self.vocab.lookup_indices(a_tokenized)
        a_tokens.extend(a_indices)

        # Add end token
        q_tokens.append(self.vocab["<end>"])
        a_tokens.append(self.vocab["<end>"])

        # for tok in tokenizer(question):
        #     q_tokens.append(self.vocab(tok))
        # for tok in tokenizer(answer):
        #     a_tokens.append(self.vocab(tok))
        # q_tokens.append(self.vocab["<end>"])
        # a_tokens.append(self.vocab["<end>"])

        return torch.tensor(q_tokens, dtype= torch.long), torch.tensor(a_tokens, dtype = torch.long)

vocab_dataset = MathDataset(dataset, split="train")
```

```python
import torch.nn as nn
import torch.nn.functional as F

class TransformerModel(nn.Module):
    def __init__(self, num_tokens, d_model, nhead, num_encoder_layers, num_decoder_layers, dim_feedforward, max_seq_length, dropout = 0.1):

        # Create transformer model, add embedding, normalization, Linear output
        super(TransformerModel, self).__init__()
        self.embedding = nn.Embedding(num_tokens, d_model)
        self.positional_encoding = nn.Parameter(torch.zeros(1, max_seq_length, d_model))
        self.transformer = nn.Transformer(
            d_model = d_model,
            nhead = nhead,
            num_encoder_layers = num_encoder_layers,
            num_decoder_layers = num_decoder_layers,
            dim_feedforward = dim_feedforward,
            dropout = dropout,
            batch_first = True
        )
        self.layer_norm = nn.LayerNorm(d_model)
        self.out = nn.Linear(d_model, num_tokens)
        self.max_seq_length = max_seq_length

    def forward(self, src, tgt):

        # Ensure correct sizes
        src_seq_length = src.size(1)
        tgt_seq_length = tgt.size(1)

        # Use layer norm w/embedding and positional encoding
        src = self.layer_norm(self.embedding(src) + self.positional_encoding[:, :src_seq_length, :])
        tgt = self.layer_norm(self.embedding(tgt) + self.positional_encoding[:, :tgt_seq_length, :])

        # Use mask for correct sizing
        tgt_mask = self.transformer.generate_square_subsequent_mask(tgt.size(1)).to(tgt.device)
        tgt_mask = tgt_mask.to(torch.bool)

        # Pass through transformer and return utput
        output = self.transformer(src, tgt, tgt_mask= tgt_mask)
        output = self.out(output)

        return output
```

```python
def evaluate(model, data_loader):

    model.eval()
    total_loss = 0

    # Start evaluation
    with torch.no_grad():

        for src, tgt in data_loader:

            src, tgt = src.to(device), tgt.to(device)

            # Take out input/output from tgt
            tgt_input = tgt[:, :-1]
            tgt_output = tgt[:, 1:]

            # Pass to model, use loss_fn
            output = model(src, tgt_input)
            loss = loss_fn(output.reshape(-1, model.out.out_features), tgt_output.reshape(-1))
            total_loss += loss.item()

    return total_loss / len(data_loader)
```

```python
from torch.nn.utils.rnn import pad_sequence

# Collate function to ensure batches are padded and have good lengths
def collate_fn(batch, vocab):

    src_batch, tgt_batch = zip(*batch)

    src_batch = pad_sequence(src_batch, padding_value = vocab["<pad>"])
    tgt_batch = pad_sequence(tgt_batch, padding_value = vocab["<pad>"])

    return src_batch.transpose(0, 1), tgt_batch.transpose(0, 1)
```

```python
from tqdm import tqdm
from torch.optim.lr_scheduler import ReduceLROnPlateau


loss_fn = nn.CrossEntropyLoss()

# Optimizer and scheduler, to lower training when it slows down
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=1e-4)
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2, verbose=True)

# Record best loss and best model
best_val_loss = float('inf')
best_model_state = None

# Use progressively increasing data
lower_bound = 5000
upper_bound = 70000
step = 5000

# 3 epochs min, slows down
epochs_per_step = 3

for num_samples in range(lower_bound, upper_bound + 1, step):
    print(f"\n--- Training on {num_samples} samples ---")

    # Create datasets with size
    full_train_dataset = MathDataset(dataset, split="train", vocab=shared_vocab, max_samples=num_samples)

    # Split train and val
    train_size = int(0.9 * len(full_train_dataset))
    val_size = len(full_train_dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(full_train_dataset, [train_size, val_size])

    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True,
                                collate_fn=lambda x: collate_fn(x, shared_vocab))
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False,
                            collate_fn=lambda x: collate_fn(x, shared_vocab))

    step_best_val_loss = float('inf')

    # Train for fewer epochs with larger datasets
    adjusted_epochs = max(1, epochs_per_step - (num_samples // 10000))

    for epoch in range(adjusted_epochs):
        # Training loop
        model.train()
        train_loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{adjusted_epochs}", leave=False)
        total_train_loss = 0

        for src, tgt in train_loop:
            src, tgt = src.to(device), tgt.to(device)
            optimizer.zero_grad()

            # Get input and output
            tgt_input = tgt[:, :-1]
            tgt_output = tgt[:, 1:]

            # Use model and back prop
            output = model(src, tgt_input)
            loss = loss_fn(output.reshape(-1, model.out.out_features), tgt_output.reshape(-1))
            loss.backward()

            # Gradient clipping to prevent exploding gradients
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            optimizer.step()
            total_train_loss += loss.item()
            train_loop.set_postfix(loss=loss.item())

        # Calculate average train loss
        avg_train_loss = total_train_loss / len(train_loader)

        # Evaluate
        val_loss = evaluate(model, val_loader)
        print(f"Samples: {num_samples}, Epoch: {epoch+1}, Train Loss: {avg_train_loss:.4f}, Val Loss: {val_loss:.4f}")

        # If necessary, slow learning rate
        scheduler.step(val_loss)

        # Early stopping check
        if val_loss < step_best_val_loss:
            step_best_val_loss = val_loss

    # Update best model if validation loss improved
    if step_best_val_loss < best_val_loss:
        best_val_loss = step_best_val_loss
        best_model_state = model.state_dict().copy()
        torch.save(model, f'model_samples_{num_samples}.pth')
        print(f"New best model saved at {num_samples} samples with loss: {best_val_loss:.4f}")
```
